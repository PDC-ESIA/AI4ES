```python
def gerar_grafico(nome_arquivo, titulo):
    import pandas as pd
    import matplotlib.pyplot as plt

    # Ler arquivo
    df = pd.read_excel(nome_arquivo)

    # Forçar nomes das colunas
    df.columns = ['Nome', 'Frequência', 'Tipo']

    # Ordenar pela frequência (maior para menor)
    df = df.sort_values(by="Frequência", ascending=False)

    # Criar gráfico no mesmo estilo
    plt.figure(figsize=(10, 5))
    plt.bar(df["Nome"], df["Frequência"])

    plt.title(titulo)
    plt.xlabel("Nome")
    plt.ylabel("Frequência")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
gerar_grafico("modelos.xlsx", "Frequência por Modelo")
```

```python
gerar_grafico("benchmarks.xlsx", "Frequência por benchmarks ")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def gerar_graficos_por_tipo(nome_arquivo):

    # Ler planilha
    df = pd.read_excel(nome_arquivo)
    df.columns = ['Nome', 'Frequência', 'Tipo']
    df['Tipo'] = df['Tipo'].str.strip()

    # Definição dos grupos
    grupo1 = ['construção/codificação', 'arquitetura', 'design', 'devOps']
    grupo2 = ['manutenção', 'testes', 'requisitos', 'Gestão']

    cores = {
        'construção/codificação': '#1f77b4',
        'arquitetura': '#ff7f0e',
        'design': '#761f75',
        'devOps': '#d62728',
        'manutenção': '#2ca02c',
        'testes': '#faff00',
        'requisitos': '#17becf',
        'Gestão': '#7f7f7f'
    }

    # ========== GRÁFICO 1 ==========
    plt.figure(figsize=(12,6))

    for tipo in grupo1:
        subset = df[df['Tipo'] == tipo]
        plt.bar(subset['Nome'], subset['Frequência'],
                color=cores[tipo], label=tipo)

    plt.title("Ferramentas - Desenvolvimento & Arquitetura")
    plt.xlabel("Ferramentas")
    plt.ylabel("Frequência")
    plt.xticks(rotation=45, ha="right")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # ========== GRÁFICO 2 ==========
    plt.figure(figsize=(12,6))

    for tipo in grupo2:
        subset = df[df['Tipo'] == tipo]
        plt.bar(subset['Nome'], subset['Frequência'],
                color=cores[tipo], label=tipo)

    plt.title("Ferramentas - Qualidade & Gestão")
    plt.xlabel("Ferramentas")
    plt.ylabel("Frequência")
    plt.xticks(rotation=45, ha="right")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
gerar_graficos_por_tipo("ferramentas.xlsx")
```
