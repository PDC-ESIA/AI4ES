# Sinteze dos Benefícios

# Sinteze dos benefícios em classificações gerais encontradas

```python
import pandas as pd
```

```python
dados_beneficios_gerais = pd.read_csv('Sinteze - analise_beneficios_gerais.csv')
df_beneficios_gerais = pd.DataFrame(dados_beneficios_gerais)
```

```python
df_beneficios_gerais
```

## Gráfico de pizza da distribuição dos benefícios gerais encontrados nos estudos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Configuração de cores e estilo
cores = plt.get_cmap('Paired')(np.linspace(0, 1, len(df_beneficios_gerais)))

plt.figure(figsize=(12, 7))

#Criando o gráfico
fatias, textos, autotextos = plt.pie(
    df_beneficios_gerais['Total'],
    labels=None,
    autopct='%1.1f%%',
    startangle=140,
    colors=cores,
    pctdistance=0.85,
    explode=[0.03] * len(df_beneficios_gerais)
)

#Ajustando a fonte das porcentagens para ficarem legíveis
plt.setp(autotextos, size=9, weight="bold", color="black")

#Adicionando legenda lateral para não poluir o gráfico
plt.legend(
    fatias,
    df_beneficios_gerais['Benefícios'],
    title="Benefícios",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.title('Distribuição de Benefícios', fontsize=15, pad=20)
plt.tight_layout()
plt.show()
```
